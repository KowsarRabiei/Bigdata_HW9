```python
from elasticsearch import Elasticsearch
from elasticsearch.helpers import scan
import pandas as pd

es = Elasticsearch(host='localhost', port=9200)

def get_data_from_elastic():
    query = {
        "query": {
            "bool": {
                "must": [
                    {
                        "exists": {
                            "field": "children"
                        }
                    },
                    {
                        "regexp": {
                            "children": ".+"
                        }
                    }
                ]
            }
        }
    }

    rel = scan(
        client=es,
        query=query,
        scroll='1m',
        index='books_info',
        raise_on_error=True,
        preserve_order=False,
        clear_scroll=True
    )

    result = list(rel)

    temp = []

    for hit in result:
        temp.append(hit['_source'])

    df_parent = pd.DataFrame(temp)

    return df_parent

df_parent = get_data_from_elastic()

print(df_parent.head())
df_parent
```

```python
from elasticsearch import Elasticsearch
from elasticsearch.helpers import scan
import pandas as pd

es = Elasticsearch(host='localhost', port=9200)

def get_data_from_elastic():
    query = {
        "query": {
            "bool": {
                "must": [
                    {
                        "exists": {
                            "field": "parent"
                        }
                    },
                    {
                        "regexp": {
                            "parent": ".+"
                        }
                    }
                ]
            }
        }
    }

    rel = scan(
        client=es,
        query=query,
        scroll='1m',
        index='books_info',
        raise_on_error=True,
        preserve_order=False,
        clear_scroll=True
    )

    result = list(rel)

    temp = []

    for hit in result:
        temp.append(hit['_source'])

    df_children = pd.DataFrame(temp)

    return df_children

df_children = get_data_from_elastic()

print(df_children.head())
df_children
```

```python
df_children['date'] = df_children['date'].str.extract(r'^(\d{4})')
df_parent['date'] = df_parent['date'].str.extract(r'^(\d{4})')
```

```python
from cassandra.cluster import Cluster
from cassandra.query import BatchStatement

# Connect to Cassandra
cluster = Cluster(['localhost'])
session = cluster.connect()

# Create keyspace and use it
session.execute("CREATE KEYSPACE IF NOT EXISTS books_info WITH REPLICATION = {'class': 'SimpleStrategy', 'replication_factor': 1}")
session.execute("USE books_info")

# Create table in Cassandra
query = "CREATE TABLE IF NOT EXISTS books_children1 (trove_id text PRIMARY KEY, title text, contributors text, url text,date text, format text)"
session.execute(query)

# Prepare the insert statement
prepared = session.prepare("INSERT INTO books_children1 (trove_id, title, contributors, url,date, format) VALUES (?, ?, ?, ?, ?,?)")

# Divide the DataFrame into smaller batches
batch_size = 100
num_batches = len(df_children) // batch_size + 1

for i in range(num_batches):
    batch = BatchStatement()
    start_index = i * batch_size
    end_index = (i + 1) * batch_size
    for index, row in df_children.iloc[start_index:end_index].iterrows():
        batch.add(prepared, (row['trove_id'], row['title'], row['contributors'], row['url'],row['date'], row['format']))
    session.execute(batch)
```

```python
# Connect to Cassandra
cluster = Cluster(['localhost'])
session = cluster.connect()

# Use the keyspace
session.execute("USE books_info")

# Select all rows from the table
query = "SELECT * FROM books_children1"
result = session.execute(query)

# Print the rows
for row in result:
    print(row)
```
